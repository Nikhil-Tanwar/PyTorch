## Reshaping, stacking, squeezing and unsqueezing tensors

* dim 0 : Collapses rows : give one value per column
* dim 1 - Collapses columns - gives one value per row


* view is like reshape - but its reference rather than pass by value : z = x.view(9, 1)

```python
x.reshape(5,2)
```

```python
x.unsqueeze(dim=0)
```

```python
x.squeeze(dim = 0)
```

* vstack - stack as rows
* hstack - stack as columns

* hstack and vstack : stacks along existing dim
* stack creates a new dim

eg: torch(2,3)
* vstack - (4,3)
* stack (dim=0)- (2, 2, 3)

```python
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
```

```python
## Permute - returns 'view' of tensor with dims permuted. (rearranged)
x = torch.randn(2, 3, 5)
x.size()
```

```python
torch.permute(x, (2, 0, 1)).shape
```

* Rand - samples from uniform distribution, [0,1), Mean - 0.5, equal prob of being selected, Img Normalisation, Scaling, Random Dropouts prob
* Randn - samples from bell or Gaussian distribtion, (-inf,+inf), Mean - 0, std dev = 1, Weight Initialisation

```python
x_original = torch.rand(224, 224, 3)
x_original.shape
```

```python
x_original = x_original.permute(2, 0 ,1)
x_original.shape
```

* Default dtype of Numpy is float64
* Default dtype of Tensor is float32

```python
# when converting from numpy to pytorch, pytorch reflects numpy default dtype (float64), unless specified
# changing the array, does not change tensor & vice versa
# Numpy to Tensor
array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)
array, tensor
```

```python
array.dtype, tensor.dtype, torch.arange(1.0, 8.0).dtype
```

```python
# Tensor to Numpy array
tensor1 = torch.arange(1.0, 8.0)
array1 = tensor.numpy()
tensor1, array1
```

```python
# Reproducibility - trying to reproduce randomness
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
t1 = torch.rand(size=(3,3))
torch.manual_seed(RANDOM_SEED)
t2 = torch.rand(size=(3,3))
t1, t2, t1 == t2
```
